```python
from  modulekmeans import Kmeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA

   
# Build a toy dataset with make_blobs
nb_clusters_dataset = 5
nb_samples = 700
nb_features = 2
X, y, centers = make_blobs(n_samples=nb_samples, centers=nb_clusters_dataset, n_features=nb_features, return_centers=True)

kmeans = Kmeans(K=5, random_state=42, tolerance=1.e-4)

kmeans.fit(X)


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plotando os clusters
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=kmeans.labels, cmap='tab10', s=50)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Clusters visualized with PCA")
plt.legend()
#plt.colorbar(label='Cluster')
plt.show()
```
